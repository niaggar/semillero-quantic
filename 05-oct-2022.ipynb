```python
import numpy as np
import numpy.random as npr
import matplotlib.pyplot as plt

WALKS = 1000
STEPS = 100
time = np.linspace(0, 1, STEPS)

robots = {}
for i in range(WALKS):
    robots[f'path-{i}'] = npr.normal(0, 1, STEPS)
    # globals()[f'path-{i}'] = npr.normal(0, 1, 100)


for walk in list(robots.values()):
    plt.plot(time, walk)

# for i in range(WALKS):
#     path = globals()[f'path-{i}']
#     plt.plot(time, path)


plt.show()



```

```python
average_steps = [0 for i in range(STEPS)]
# print(average_steps)

for i in range(WALKS):
    for j in range(STEPS):
        average_steps[j] += robots[f'path-{i}'][j]

average_steps = np.array(average_steps) / WALKS

# plt.plot(time, robots['path-1'])
# plt.plot(time, robots['path-89'])
for walk in list(robots.values()):
    plt.plot(time, walk)
plt.plot(time, average_steps, "navy", lw = 4 )
plt.show()
```
