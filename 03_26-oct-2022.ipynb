```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def metodo_de_euler(ecu, x0, xf, y0, dx):
    x = np.arange(x0, xf, dx)
    y = np.zeros(len(x))
    y[0] = y0

    for i in range(1, len(x)):
        y[i] = y[i-1] + dx * ecu(x[i-1], y[i-1])

    return x, y
```

```python
def metodo_de_euler_mejorado(ecu, x0, xf, y0, dx):
    x = np.arange(x0, xf, dx)
    y = np.zeros(len(x))
    y[0] = y0

    for i in range(1, len(x)):
        u = y[i-1] + dx * ecu(x[i-1], y[i-1])
        y[i] = y[i-1] + dx/2 * (ecu(x[i-1], y[i-1]) + ecu(x[i], u)) 

    return x, y
```

```python
def Ec_differencial(x, y):
    dydx = ((x**2) * (y)) + 3 * x
    return dydx


resultado_euler = metodo_de_euler(Ec_differencial, 0, 2, 1, 0.1)
resultado_euler_mejorado = metodo_de_euler_mejorado(Ec_differencial, 0, 2, 1, 0.1)


plt.plot(resultado_euler[0], resultado_euler[1], label="Euler")
plt.plot(resultado_euler_mejorado[0], resultado_euler_mejorado[1], label="Euler Mejorado")
plt.legend()
plt.rcParams['figure.figsize'] = [15, 10]
plt.show()




```
