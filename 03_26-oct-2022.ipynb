```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def metodo_de_euler(ecu, x0, xf, y0, dx):
    x = np.arange(x0, xf, dx)
    y = np.zeros(len(x))
    y[0] = y0

    for i in range(1, len(x)):
        y[i] = y[i-1] + dx * ecu(x[i-1], y[i-1])

    return x, y
```

```python
def metodo_de_euler_mejorado(ecu, x0, xf, y0, dx):
    x = np.arange(x0, xf, dx)
    y = np.zeros(len(x))
    y[0] = y0

    for i in range(1, len(x)):
        u = y[i-1] + dx * ecu(x[i-1], y[i-1])
        y[i] = y[i-1] + dx/2 * (ecu(x[i-1], y[i-1]) + ecu(x[i], u)) 

    return x, y
```

```python
def Ec_differencial1(x, y):
    dydx = ((x**2) * (y)) + 3 * x
    return dydx

def Ec_differencial2(x, y):
    dydx = 2 * x * y
    return dydx

def Ec_differencial3(x, y):
    dydx = np.sin(x) + np.cos(y)
    return dydx


resultado_euler = metodo_de_euler(Ec_differencial1, -2, 2, 0, 0.1)
resultado_euler_mejorado = metodo_de_euler_mejorado(Ec_differencial1, -2, 2, 0, 0.1)

plt.plot(resultado_euler[0], resultado_euler[1], label="Euler")
plt.plot(resultado_euler_mejorado[0], resultado_euler_mejorado[1], label="Euler Mejorado")
plt.legend()
plt.rcParams['figure.figsize'] = [15, 10]
plt.show()


resultado_euler = metodo_de_euler(Ec_differencial2, 0, 2, 1, 0.1)
resultado_euler_mejorado = metodo_de_euler_mejorado(Ec_differencial2, 0, 2, 1, 0.1)

plt.plot(resultado_euler[0], resultado_euler[1], label="Euler")
plt.plot(resultado_euler_mejorado[0], resultado_euler_mejorado[1], label="Euler Mejorado")
plt.legend()
plt.rcParams['figure.figsize'] = [15, 10]
plt.show()


resultado_euler = metodo_de_euler(Ec_differencial3, 0, np.pi * 4, 1, 0.1)
resultado_euler_mejorado = metodo_de_euler_mejorado(Ec_differencial3, 0, np.pi * 4, 1, 0.1)

plt.plot(resultado_euler[0], resultado_euler[1], label="Euler")
plt.plot(resultado_euler_mejorado[0], resultado_euler_mejorado[1], label="Euler Mejorado")
plt.legend()
plt.rcParams['figure.figsize'] = [15, 10]
plt.show()
```

```python
# Directional field
def campo_de_direcciones(ecu, x0, xf, y0, yf, dx, dy):
    x = np.arange(x0, xf, dx)
    y = np.arange(y0, yf, dy)

    X, Y = np.meshgrid(x, y)
    U = 1
    V = ecu(X, Y)
    V = V / np.sqrt(U**2 + V**2)

    # Graficar ejes de coordenadas
    plt.plot([x0, xf], [0, 0], color='black')
    plt.plot([0, 0], [y0, yf], color='black')
    
    plt.quiver(X, Y, U, V)
    plt.show()



# Directional field
def directional_field(ecu, x0, xf, y0, yf, dx, dy):
    x = np.arange(x0, xf, dx)
    y = np.arange(y0, yf, dy)

    X, Y = np.meshgrid(x, y)
    U = 1
    V = ecu(X, Y)
    V = V / np.sqrt(U**2 + V**2)

    resultado_euler = metodo_de_euler(ecu, -2, xf, 0, dx)
    resultado_euler_mejorado = metodo_de_euler_mejorado(ecu, -2, xf, 0, dx)

    plt.plot(resultado_euler[0], resultado_euler[1], label="Euler")
    plt.plot(resultado_euler_mejorado[0], resultado_euler_mejorado[1], label="Euler Mejorado")
    plt.legend()
    plt.rcParams['figure.figsize'] = [15, 10]

    # Graficar ejes de coordenadas
    plt.plot([x0, xf], [0, 0], color='black')
    plt.plot([0, 0], [y0, yf], color='black')

    plt.ylim(y0, yf)
    plt.xlim(x0, xf)
    
    plt.quiver(X, Y, U, V)
    plt.show()

directional_field(Ec_differencial1, -2, 1.2, -30, 0, 0.1, 0.1)
```
